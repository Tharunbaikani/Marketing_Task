```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
transactions = pd.read_csv('transaction_data.csv')
behavior = pd.read_csv('purchase_behaviour.csv')
```

```python
# Merge datasets on loyalty card number
merged_data = pd.merge(transactions, behavior, on='LYLTY_CARD_NBR', how='left')
```

```python
# Clean data
merged_data['DATE'] = pd.to_datetime(merged_data['DATE'])
```

```python
# Calculate product profitability (assuming TOT_SALES = revenue)
product_profit = merged_data.groupby(['PROD_NBR', 'PROD_NAME'])['TOT_SALES'].sum().reset_index()
top_products = product_profit.sort_values('TOT_SALES', ascending=False).head(3)
```

```python
# RFM Analysis for Loyal Customers
snapshot_date = merged_data['DATE'].max() + pd.DateOffset(days=1)
rfm = merged_data.groupby('LYLTY_CARD_NBR').agg({
    'DATE': lambda x: (snapshot_date - x.max()).days,  # Recency
    'TXN_ID': 'nunique',                               # Frequency
    'TOT_SALES': 'sum'                                 # Monetary
}).reset_index()
rfm.columns = ['LYLTY_CARD_NBR', 'Recency', 'Frequency', 'Monetary']
```

```python
# Calculate RFM score (lower recency = better, higher frequency/monetary = better)
rfm['R_rank'] = rfm['Recency'].rank(ascending=False)   # Recent customers rank higher
rfm['F_rank'] = rfm['Frequency'].rank(ascending=True)
rfm['M_rank'] = rfm['Monetary'].rank(ascending=True)
rfm['RFM_Score'] = (rfm['R_rank'] + rfm['F_rank'] + rfm['M_rank']) / 3
```

```python
# Identify top 20% loyal customers
loyal_customers = rfm[rfm['RFM_Score'] >= rfm['RFM_Score'].quantile(0.8)]
```

```python
# Merge with behavioral data to analyze demographics
loyal_profile = pd.merge(loyal_customers, behavior, on='LYLTY_CARD_NBR', how='left')
```

```python
# Analyze characteristics
lifestage_dist = loyal_profile['LIFESTAGE'].value_counts(normalize=True)
premium_dist = loyal_profile['PREMIUM_CUSTOMER'].value_counts(normalize=True)
```

```python
# Results
print("Top 3 Most Profitable Products:")
print(top_products[['PROD_NAME', 'TOT_SALES']])
```

```python
print("\nCharacteristics of Loyal Customers:")
print(f"Lifestage Distribution:\n{lifestage_dist}")
print(f"\nPremium vs Budget Buyers:\n{premium_dist}")
```

```python
# Visualizations
plt.figure(figsize=(10, 5))
sns.barplot(x='PROD_NAME', y='TOT_SALES', data=top_products)
plt.title('Top 3 Profitable Products by Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=loyal_profile, x='LIFESTAGE', hue='PREMIUM_CUSTOMER')
plt.title('Lifestage and Premium Status of Loyal Customers')
plt.xticks(rotation=45)
plt.show()
```
